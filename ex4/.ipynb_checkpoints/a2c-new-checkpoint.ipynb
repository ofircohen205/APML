```python
import torch
import torch.nn.functional as F
import torch.optim as optim
import torch.nn as nn

import torch
import random
import gym
import numpy as np
from torch.utils.tensorboard import SummaryWriter
from torch.distributions import Categorical
from tqdm import trange

from base_policy import BasePolicy
from snake import THE_DEATH_PENALTY
from train_snake import *

def conv2d_size_out(size, kernel_size=3, stride=1):
    return (size - (kernel_size - 1) - 1) // stride + 1


class ActorCriticModel(nn.Module):
    def __init__(self, h=9, w=9, inputs=10, outputs=3):
        """
        :param h:
        :param w:
        :param outputs: number of actions (3)
        """
        super(ActorCriticModel, self).__init__()
        self.conv1 = nn.Conv2d(inputs, 16, kernel_size=3, stride=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.conv3 = nn.Conv2d(32, 32, kernel_size=3, stride=1)
        self.bn3 = nn.BatchNorm2d(32)

        convw = conv2d_size_out(conv2d_size_out(conv2d_size_out(w)))
        convh = conv2d_size_out(conv2d_size_out(conv2d_size_out(h)))
        linear_input_size = convw * convh * 32
        
#         self.critic_fc = nn.Linear(linear_input_size, linear_input_size // 2)
#         self.critic_head = nn.Linear(linear_input_size // 2, outputs)
        
#         self.actor_fc = nn.Linear(linear_input_size, linear_input_size // 2)
#         self.actor_head = nn.Linear(linear_input_size // 2, outputs)
        self.critic_head = nn.Linear(linear_input_size, outputs)
        self.actor_head = nn.Linear(linear_input_size, outputs)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)  # nhwc -> nchw
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.relu(self.bn3(self.conv3(x)))
        x = x.view(x.size(0), -1)
        
#         critic_x = F.relu(self.critic_fc(x))
#         q_values = self.critic_head(critic_x)
        
#         actor_x = F.relu(self.actor_fc(x))
#         policy_dist = F.softmax(self.actor_head(actor_x), dim=-1)
        
        q_value = self.critic_head(x)
        policy_dist = F.softmax(self.actor_head(x), dim=-1)
        return q_value, policy_dist
```

```python
class ActorCriticPolicy(BasePolicy):
    def __init__(self, gamma, max_norm, model,
                 action_space: gym.Space, summery_writer: SummaryWriter,
                 normalized_rewards, optimizer, lr):
        super(ActorCriticPolicy, self).__init__(0, gamma, model, action_space, summery_writer, lr)
        self.normalized_rewards = normalized_rewards
        self.optimizer = getattr(optim, optimizer)(self.model.parameters(), lr=lr)
        if max_norm > 0:
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm)
        # memmory buffer
        self.probs = []
        self.actions = []
        self.q_values = []
        self.log_probs = []
        self.rewards = []
        self.entropy_term = torch.FloatTensor([0])

    def select_action(self, state, epsilon=None, global_step=None, alpha=None):
        value, policy_dist = self.model(state)
        return value, policy_dist

    def optimize(self, global_step, alpha=None):        
        if self.normalized_rewards:
            normalized_rewards = np.asarray(self.rewards)
            normalized_rewards = (normalized_rewards - normalized_rewards.mean()) / (normalized_rewards.std() + 1e-8)
        else:
            normalized_rewards = self.rewards
        
        # convert episode memmory to torch tensors
        q_values = torch.cat(self.q_values)
        log_probs = torch.cat(self.log_probs)
        actions = torch.cat(self.actions)
        probs = torch.cat(self.probs)
        
        # get V (expectation of q values)
        values = (q_values * probs).sum(dim=1) 
        # get Q estimation
        done_state_mask = torch.FloatTensor(self.rewards) != -5
        q_val_estimates = torch.FloatTensor(normalized_rewards) + (done_state_mask * self.gamma * values)
        q_val_estimates = q_val_estimates.view(-1,1)
        # get Q values for the sampled actions
        q_values = q_values.gather(1, actions.view(-1,1))

        # calculate loss
        advantage = q_values - values
        actor_loss = (log_probs * advantage.detach()).mean()
        critic_loss = F.mse_loss(target=q_val_estimates.detach(), input=q_values)
        # minimize critic loss and maximize actor loss
        loss = critic_loss - (actor_loss + alpha * (self.entropy_term / len(self.rewards)))

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        return loss.item()

    def clear_buffer(self):
        self.probs = []
        self.actions = []
        self.q_values = []
        self.log_probs = []
        self.rewards = []
        self.entropy_term = torch.FloatTensor([0])
```

```python
def train_policy(steps, lr, max_alpha, max_norm, gamma, max_episodes, normalized_rewards, optimizer, log_dir):
    game = SnakeWrapper()
    writer = SummaryWriter(log_dir=log_dir)
    a2c_model = ActorCriticModel(outputs=game.action_space.n)
    policy = ActorCriticPolicy(gamma=gamma, max_norm=max_norm,
                               model=a2c_model, action_space=game.action_space, summery_writer=writer,
                               lr=lr, optimizer=optimizer, normalized_rewards=normalized_rewards)


    prog_bar = trange(max_episodes, desc='', leave=True)
    for episode in prog_bar:
        state = game.reset()
        alpha = max_alpha * math.exp(-1. * episode / (max_episodes / 2))
        writer.add_scalar('training/alpha', alpha, episode)

        for step in range(steps):
            q_value, policy_dist = policy.select_action(torch.FloatTensor(state))
            dist = Categorical(probs=policy_dist)
            action = dist.sample()
            log_prob = dist.log_prob(value=action)
            entropy = dist.entropy()
            
            new_state, reward = game.step(action.item())
            
            # add records
            policy.probs.append(policy_dist)
            policy.log_probs.append(log_prob)
            policy.rewards.append(reward)
            policy.q_values.append(q_value)
            policy.actions.append(action)
            policy.entropy_term += entropy
            
            state = new_state
            
            if reward == THE_DEATH_PENALTY:
                break

        loss = policy.optimize(global_step=episode, alpha=alpha)
        
        writer.add_scalar('training/episode_length', step, episode)
        writer.add_scalar('training/episode_rewards', np.sum(policy.rewards), episode)
        writer.add_scalar('training/loss', loss, episode)
        prog_bar.set_description(f"episode: {episode+1}, loss: {loss:.3f} "
                                 f"total reward: {np.sum(policy.rewards)}, steps: {step}")
        policy.clear_buffer()

#     torch.save({'model_state_dict': policy.model.state_dict()}, f'vanilla_reinforce.pkl')
    writer.close()
    return policy
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

def test_policy(policy, iterations, render):
    game = SnakeWrapper()
    action_space = SnakeWrapper.action_space
    state = game.reset()
    if render:
        game.render()
            
    total_rewards = []
    steps_survived = []
    
    total = 0
    last = 0
    for i in range(iterations):
        state_tensor = torch.FloatTensor(state)
        _, policy_dist = policy.select_action(torch.FloatTensor(state))
        action = Categorical(probs=policy_dist).sample().item()
        state, reward = game.step(action)
        if reward < 0:
            steps_survived.append(i - last)
            total_rewards.append(total)
            total = 0
            last = i
        else:
            total += reward
        if render:
            game.render()
            
    # plot reward histogram      
    plt.figure()
    plt.title("reward histogram")
    plt.hist(x=total_rewards, bins=100)
    plt.xlabel("reward")
    plt.ylabel("count")
    plt.show()

    # plot steps histogram  
    plt.figure()
    plt.title("steps histogram")
    plt.hist(x=steps_survived, bins=100)
    plt.xlabel("# of steps")
    plt.ylabel("count")
    plt.show()
    
    # plot 2d histogram  
    plt.figure()
    plt.title("step vs rewards histogram")
    plt.hist2d(x=steps_survived, y=total_rewards, bins=100)
    plt.xlabel("# of steps")
    plt.ylabel("reward")
    plt.show()
```

```python
policy = train_policy(steps=1000, lr=0.01, max_alpha=0.3, max_norm=0, 
                      normalized_rewards=False, optimizer="Adam",
                      gamma=0.01, max_episodes=1000,
                      log_dir=r"./logs/a2c/")
```

```python
# 2nd run
policy.eval()
test_policy(policy, iterations=10000, render=False)
```
